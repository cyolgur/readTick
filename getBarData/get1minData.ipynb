```python
PATH = 'C:/tickdata_checkOrder/'
fileList = os.listdir('C:/tickdata_checkOrder')


def readTickDataRaw(path):
    tmp = pd.read_csv(path,
                      usecols=['TDATE', 'TTIME', 'UPDATEMILLISEC', 'LASTPX', 'HIGHPX', 'LOWPX',
                               'TQ', 'TM', 'RISELIMIT', 'FALLLIMIT', 'S1', 'B1', 'SV1', 'BV1',
                               'OPENINTEREST'],
                      dtype={
                          'TDATE': str,
                          'TTIME': str,
                          'UPDATEMILLISEC': str,
                          'LASTPX': np.float64,
                          'HIGHPX': np.float64,
                          'LOWPX': np.float64,
                          'TQ': np.float64,
                          'TM': np.float64,
                          'RISELIMIT': np.float64,
                          'FALLLIMIT': np.float64,
                          'S1': np.float64,
                          'B1': np.float64,
                          'SV1': np.float64,
                          'BV1': np.float64,
                          'OPENINTEREST': np.float64,
                      },
                      parse_dates=[['TDATE', 'TTIME', 'UPDATEMILLISEC']],
                      engine='c')
    tmp['TDATE_TTIME_UPDATEMILLISEC'] = pd.to_datetime(tmp['TDATE_TTIME_UPDATEMILLISEC'], format='%Y%m%d %H%M%S %f')
    tmp['B1'] = np.where((tmp['B1'] < tmp['FALLLIMIT']) | (tmp['B1'] > tmp['RISELIMIT']),
                         tmp['FALLLIMIT'],
                         tmp['B1'])
    tmp['S1'] = np.where((tmp['S1'] < tmp['FALLLIMIT']) | (tmp['S1'] > tmp['RISELIMIT']),
                         tmp['RISELIMIT'],
                         tmp['S1'])
    tmp.drop(columns=['RISELIMIT', 'FALLLIMIT'], inplace=True)
    tmp.set_index(['TDATE_TTIME_UPDATEMILLISEC'], inplace=True)
    return tmp
```

```python
DATA_SRC = 'D:/data_1min_consecutive'
SYMBOLS = ['AG', 'AL', 'AU', 'BU', 'C', 'CU', 'EB', 'EG', 'FG', 'FU',
           'HC', 'I', 'J', 'JM', 'L', 'M', 'MA', 'NI', 'OI', 'P', 'PF', 'PP',
           'RB', 'RM', 'RU', 'SA', 'SC', 'SP', 'TA', 'V', 'Y', 'ZC', 'ZN']


def get_file_path(inst, date_str):
    return os.path.join(DATA_SRC, inst, '{}.csv'.format(date_str))


def read_file(file_path):
    data = pd.read_csv(file_path,
                       index_col=[0,])
    return data


def handle_data(data):
    data['TradeDate'] = data['TradeDate'].apply(lambda a_str: datetime.datetime.strptime(a_str, '%Y-%m-%d').date())
    data['TradeTime'] = data['TradeTime'].apply(lambda a_str: datetime.datetime.strptime(a_str, '%H:%M:%S').time())
    # 这里到底要不要删除 那些没有的夜盘数据呢？
    return data


def read_consecutive_data(inst, date_str):
    data = read_file(get_file_path(inst, date_str))
    return handle_data(data)
```

```python
STARTTIME1 = datetime.time(21, 0, 0)
ENDTIME1 = datetime.time(2, 30, 0)
STARTTIME2 = datetime.time(9, 0, 0)
ENDTIME2 = datetime.time(11, 30, 0)
STARTTIME3 = datetime.time(13, 30, 0)
ENDTIME3 = datetime.time(15, 0, 0)
def part_sign(a_time):
    if (a_time <= ENDTIME1) or (a_time >= STARTTIME1):
        return 1
    elif (a_time >= STARTTIME2) and (a_time <= ENDTIME2):
        return 2
    elif (a_time >= STARTTIME3) and (a_time <= ENDTIME3):
        return 3
    else:
        raise ValueError('timeError!')


# 在groupby的过程中需谨记：有的日子是没有夜盘的，也就是在groupby的过程中会少一栏1
def dowm_sample_3part(data):
    data['time_flag'] = data['TradeTime'].apply(part_sign)
    data_new = data.groupby(data['time_flag']).agg({'Code': 'first',
                                                    'TradeDate': 'first',
                                                    'OpenPrice': 'first',
                                                    'HighPrice': np.nanmax,
                                                    'LowPrice': np.nanmin,
                                                    'ClosePrice': 'last',
                                                    'TradeVolume': np.nansum,
                                                    'OpenInterest': 'last',
                                                    'Turnover': np.nansum})
    if len(data_new) == 3:
        return data_new
    else:
        data_new = data_new.reindex([1,2,3])
        data_new[['Code', 'TradeDate']] = data_new[['Code', 'TradeDate']].fillna(method='backfill')
        return data_new
```

```python
def calcTradePrice(data, tradeMinutes):
    trade_startTime = [STARTTIME1, STARTTIME2, STARTTIME3]
    trade_endTime = [datetime.time(i.hour, i.minute + tradeMinutes, 0) for i in trade_startTime]
    mask_night = (data['TradeTime'] >= trade_startTime[0]) & (data['TradeTime'] <= trade_endTime[0])
    mask_am = (data['TradeTime'] >= trade_startTime[1]) & (data['TradeTime'] <= trade_endTime[1])
    mask_pm = (data['TradeTime'] >= trade_startTime[2]) & (data['TradeTime'] <= trade_endTime[2])
    res = []
    for mask in [mask_night, mask_am, mask_pm]:
        tradePrice = data.loc[mask, 'ClosePrice']
        if tradePrice.empty:
            res.append(np.nan)
        else:
            res.append(tradePrice.mean())
    return res
```

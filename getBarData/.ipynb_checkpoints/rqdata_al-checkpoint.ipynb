```python
data = futures.get_dominant_price('FG',
                           start_date='2017-01-01',
                           end_date='2020-12-31',
                           frequency='1d',
                           fields=['open', 'close'],
                           adjust_type='pre', 
                           adjust_method='open_ratio')
```

```python
OLD_PATH = 'D:/data_new'
NEW_PATH = 'D:/data_3part_consecutive_by_name'
OPEN_NAME = 'index_open.csv'
CLOSE_NAME = 'index_close.csv'
```

```python
old_close = pd.read_csv(os.path.join(OLD_PATH, CLOSE_NAME),
                       index_col = [0,1],
                      parse_dates=True)
old_al_close = old_close['AL'].groupby(level=0).last()
```

```python
new_close = pd.read_csv(os.path.join(NEW_PATH, CLOSE_NAME),
                       index_col = [0,1],
                        parse_dates=True)
new_al_close = new_close['AL'].groupby(level=0).last()
```

```python
(new_al_close/old_al_close).plot()
```

```python
data_copy = data.copy(deep=True)
data_copy = data_copy.droplevel(level=0)
data_copy.index = pd.to_datetime(data_copy.index)
```

```python
(new_al_close/data_copy['close']).plot()
```

```python
(data_copy['close']/old_al_close).plot()
```

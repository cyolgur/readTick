```python
给定一个整数 n，计算所有小于等于 n 的非负整数中数字 1 出现的个数。
```

```python
def calculateMinimumHP(dungeon) -> int:
    if len(dungeon) == 0:
        return 1
    nrow = len(dungeon)
    ncol = len(dungeon[0])
    res = [[0 for j in range(ncol)] for i in range(nrow)]
    for i in range(nrow-1, -1, -1):
        for j in range(nrow-1, -1, -1):
            print(i,j,res)
            if (i==nrow-1) and (j==ncol-1):
                res[i][j] = dungeon[i][j]
            elif i== nrow-1:
                res[i][j] = min(res[i][j+1]+dungeon[i][j], dungeon[i][j])
            elif j==ncol-1:
                res[i][j] = min(res[i+1][j]+dungeon[i][j], dungeon[i][j])
            else:
                res[i][j] = min(max(res[i+1][j]+dungeon[i][j], res[i][j+1]+dungeon[i][j]), dungeon[i][j])
    
    if res[0][0]>=0:
        return 1
    else:
        return -res[0][0]+1
```

```python
calculateMinimumHP([[0,-3]])
```
